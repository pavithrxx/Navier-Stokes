# Step 1 LinearConvection

```python
import numpy                           
from matplotlib import pyplot as plt   

#wedefine an evenly spaced grid of points within a spatial domain that is 2 units of length wide, i.e., xi(0,2)

# Define the number of spatial grid points
nx = 41  # try changing this number from 41 to 81 

# Calculate the distance between adjacent points in the spatial grid
dx = 2 / (nx - 1)

# Define the number of timesteps to calculate
nt = 25

# Define the amount of time each timestep covers (delta t)
dt = .025

# Assume a wave speed of c = 1
c = 1

# Initialize the u array with ones
u = numpy.ones(nx)

# Set u = 2 between 0.5 and 1, as per the initial conditions
u[int(.5 / dx):int(1 / dx + 1)] = 2

# Print the initial condition array
print(u)

# Plot the initial condition
plt.plot(numpy.linspace(0, 2, nx), u)

# Initialize a temporary array for the new values of u
un = numpy.ones(nx)

# Loop over each timestep
for n in range(nt):
    un = u.copy()  # Copy the existing values of u into un
    
    # Loop over each spatial point (except the first one)
    for i in range(1, nx):
        #implement the discretization of the convection equation using a finite-difference scheme.
        # Update the value of u at each point based on the previous values
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
    
    # Plot the updated values of u at this timestep
    plt.plot(numpy.linspace(0, 2, nx), u)

# Print the final array after all timesteps
print(u)

# Display the plot with all timesteps
plt.show()
```

# Step 4- Burgers' Equation with a saw-tooth IC and periodic BCs

```python
import numpy  # for numerical operations
import sympy  # for symbolic mathematics
from sympy import init_printing  # for pretty printing
from sympy.utilities.lambdify import lambdify  # to convert sympy expressions to functions
from matplotlib import pyplot  # for plotting

### Analytical Solution
# Define the symbolic variables
x, nu, t = sympy.symbols('x nu t')

# Define the function phi for the analytical solution
phi = (sympy.exp(-(x - 4 * t)**2 / (4 * nu * (t + 1))) +
       sympy.exp(-(x - 4 * t - 2 * numpy.pi)**2 / (4 * nu * (t + 1))))

# Compute the derivative of phi with respect to x
phiprime = phi.diff(x)

# Define the analytical solution for u
u = -2 * nu * (phiprime / phi) + 4

# Convert the symbolic expression to a numerical function
ufunc = lambdify((t, x, nu), u)

### Variable declarations - Numerical Solution
nx = 101  # number of spatial grid points
nt = 10  # number of timesteps- can be experimented
dx = 2 * numpy.pi / (nx - 1)  # spatial step size
nu = .07  # viscosity
dt = dx * nu  # time step size

# Create a spatial grid
x = numpy.linspace(0, 2 * numpy.pi, nx)

# Initialize arrays
un = numpy.empty(nx)
t = 0  # initial time

# Initialize u using the analytical solution at t=0
u = numpy.asarray([ufunc(t, x0, nu) for x0 in x])

# Plot the initial condition
pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.plot(x, u, marker='o', lw=2)
pyplot.xlim([0, 2 * numpy.pi])
pyplot.ylim([0, 10])

### Numerical calculation
# Time-stepping loop
for n in range(nt):
    un = u.copy()  # copy the current state of u to un
    # Update u using the finite difference method
    for i in range(1, nx-1):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1]) + nu * dt / dx**2 * (un[i+1] - 2 * un[i] + un[i-1])
    # Apply periodic boundary conditions
    u[0] = un[0] - un[0] * dt / dx * (un[0] - un[-2]) + nu * dt / dx**2 * (un[1] - 2 * un[0] + un[-2])
    u[-1] = u[0]

# Compute the analytical solution at the final time
u_analytical = numpy.asarray([ufunc(nt * dt, xi, nu) for xi in x])

### Plotting
# Plot the numerical and analytical solutions
pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.plot(x, u, marker='o', lw=2, label='Computational')
pyplot.plot(x, u_analytical, label='Analytical')
pyplot.xlim([0, 2 * numpy.pi])
pyplot.ylim([0, 10])
pyplot.legend()

# Show the plot
pyplot.show()
```

# Step 5 Linear convection in 2D with a square-function IC and appropriate BCs.

```python
from mpl_toolkits.mplot3d import Axes3D    ## Library required for projected 3D plots
import numpy
from matplotlib import pyplot
from matplotlib import cm 
import time

### Variable declarations
nx = 81  # Number of grid points in the x-direction
ny = 81  # Number of grid points in the y-direction
nt = 100  # Number of timesteps
c = 1  # Wave speed
dx = 2 / (nx - 1)  # Distance between grid points in the x-direction
dy = 2 / (ny - 1)  # Distance between grid points in the y-direction
sigma = .2  # CFL number
dt = sigma * dx  # Timestep size, calculated based on CFL condition

x = numpy.linspace(0, 2, nx)  # x coordinates
y = numpy.linspace(0, 2, ny)  # y coordinates

u = numpy.ones((ny, nx))  # Create a 2D array of ones with size ny x nx
un = numpy.ones((ny, nx))  # Create another 2D array of ones for temporary storage

### Assign initial conditions
## Set hat function I.C.: u(.5<=x<=1 && .5<=y<=1) is 2
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2 

### Plot Initial Condition
## The figsize parameter can be used to produce different sized images
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d' to create 3D axes
X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting                          
surf = ax.plot_surface(X, Y, u[:], cmap=cm.viridis)  # Plot the surface
pyplot.show()

### Numerical Calculation with Loops
t = time.time()  # Start time for performance measurement
u = numpy.ones((ny, nx))
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2  # Re-assign initial conditions

for n in range(nt + 1):  # Loop across number of time steps
    un = u.copy()  # Copy the values of u into un
    row, col = u.shape
    for j in range(1, row):
        for i in range(1, col):
            # Apply finite difference method
            u[j, i] = (un[j, i] - (c * dt / dx * (un[j, i] - un[j, i - 1])) -
                                  (c * dt / dy * (un[j, i] - un[j - 1, i])))
    # Apply boundary conditions
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1

fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d' to create 3D axes
surf2 = ax.plot_surface(X, Y, u[:], cmap=cm.viridis)  # Plot the surface after calculation
pyplot.show()
print('The Calculation with loops took: ' + str('{0:.2f}'.format(time.time() - t)) + 's')

### Numerical Calculation with Arrays
t = time.time()  # Start time for performance measurement
u = numpy.ones((ny, nx))
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2  # Re-assign initial conditions

for n in range(nt + 1):  # Loop across number of time steps
    un = u.copy()  # Copy the values of u into un
    # Apply finite difference method using array operations
    u[1:, 1:] = (un[1:, 1:] - (c * dt / dx * (un[1:, 1:] - un[1:, :-1])) -
                              (c * dt / dy * (un[1:, 1:] - un[:-1, 1:])))
    # Apply boundary conditions
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1

fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d' to create 3D axes
surf2 = ax.plot_surface(X, Y, u[:], cmap=cm.viridis)  # Plot the surface after calculation
pyplot.show()
print('The Calculation with arrays took: ' + str('{0:.2f}'.format(time.time() - t)) + 's')
```

# Step 6: With the same IC/BCs, nonlinear convection in 2D.

```python
from mpl_toolkits.mplot3d import Axes3D  ## New Library required for projected 3D plots
import numpy
from matplotlib import pyplot
from matplotlib import cm 

### Variable declarations
nx = 81
ny = 81
nt = 100
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
sigma = .2
dt = sigma * dx

x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 2, ny)

u = numpy.ones((ny, nx))  # Create a 2D array of ones with size ny x nx for u
v = numpy.ones((ny, nx))  # Create a 2D array of ones with size ny x nx for v
un = numpy.ones((ny, nx))  # Create a temporary array for u
vn = numpy.ones((ny, nx))  # Create a temporary array for v

### Assign initial conditions
## Set hat function I.C.: u(.5<=x<=1 && .5<=y<=1) is 2
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
## Set hat function I.C.: v(.5<=x<=1 && .5<=y<=1) is 2
v[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2

### Plot Initial Condition
## The figsize parameter can be used to produce different sized images
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Create 3D axes
X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting
ax.plot_surface(X, Y, u, cmap=cm.viridis, rstride=2, cstride=2)  # Plot the initial condition
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

# Calc with Arrays
for n in range(nt + 1):  # Loop across number of time steps
    un = u.copy()  # Copy the values of u into un
    vn = v.copy()  # Copy the values of v into vn

    # Update u using the finite difference method with array operations
    u[1:, 1:] = (un[1:, 1:] - (un[1:, 1:] * c * dt / dx * (un[1:, 1:] - un[:-1, 1:])) -
                 vn[1:, 1:] * c * dt / dy * (un[1:, 1:] - un[1:, :-1]))
    
    # Update v using the finite difference method with array operations
    v[1:, 1:] = (vn[1:, 1:] - (un[1:, 1:] * c * dt / dx * (vn[1:, 1:] - vn[:-1, 1:])) -
                 vn[1:, 1:] * c * dt / dy * (vn[1:, 1:] - vn[1:, :-1]))
    
    # Apply boundary conditions for u
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1
    
    # Apply boundary conditions for v
    v[0, :] = 1
    v[-1, :] = 1
    v[:, 0] = 1
    v[:, -1] = 1

### Plot Resulting Surface
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Create 3D axes for plotting the resulting surface of u after the numerical calculation.
X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting

# Plot the resulting surface for u after the numerical calculation
ax.plot_surface(X, Y, u, cmap=cm.viridis, rstride=2, cstride=2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

pyplot.show()
```

# Step 7: With the same IC/BCs, diffusion in 2D.

```python
import numpy	
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D ## Library for 3D projection plots

### Variable declarations
nx = 31  # Number of grid points in x-direction
ny = 31  # Number of grid points in y-direction
nt = 17  # Number of time steps
nu = .05  # Viscosity
dx = 2 / (nx - 1)  # Distance between grid points in x-direction
dy = 2 / (ny - 1)  # Distance between grid points in y-direction
sigma = .25  # Stability parameter
dt = sigma * dx * dy / nu  # Time step size

x = numpy.linspace(0, 2, nx)  # x-grid points
y = numpy.linspace(0, 2, ny)  # y-grid points

u = numpy.ones((ny, nx))  # Create a 2D array of ones
un = numpy.ones((ny, nx))  # Temporary array to hold values during updates

### Assign initial conditions
# Set hat function initial condition: u(.5<=x<=1 && .5<=y<=1) is 2
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2  

### Plot Initial Condition
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d'
X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting
surf = ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap=cm.viridis,
                       linewidth=0, antialiased=False) #Plot the initial condition.

ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
ax.set_zlim(1, 2.5)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

### Run through nt timesteps
def diffuse(nt):
    # Reinitialize the initial condition inside the function
    u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2  
    
    for n in range(nt + 1): 
        un = u.copy()  # Copy the current state to temporary array
        u[1:-1, 1:-1] = (un[1:-1, 1:-1] + 
                        nu * dt / dx**2 * 
                        (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, 0:-2]) +
                        nu * dt / dy**2 * 
                        (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[0:-2, 1:-1]))
        # Apply boundary conditions
        u[0, :] = 1
        u[-1, :] = 1
        u[:, 0] = 1
        u[:, -1] = 1
    
    # Plot the result after nt timesteps
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d'
    surf = ax.plot_surface(X, Y, u[:], rstride=1, cstride=1, cmap=cm.viridis,
                           linewidth=0, antialiased=True)
    ax.set_zlim(1, 2) #experiment in range 2-2.5
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')

# Call the diffuse function to run the simulation and plot the result
diffuse(nt)
```

# Step 8: Burgers' Equation in 2D

```python
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

### Variable declarations
nx = 41  # Number of grid points in x-direction
ny = 41  # Number of grid points in y-direction
nt = 120  # Number of time steps
c = 1  # Convection coefficient
dx = 2 / (nx - 1)  # Distance between grid points in x-direction
dy = 2 / (ny - 1)  # Distance between grid points in y-direction
sigma = .0009  # Stability parameter
nu = 0.01  # Viscosity
dt = sigma * dx * dy / nu  # Time step size

x = numpy.linspace(0, 2, nx)  # x-grid points
y = numpy.linspace(0, 2, ny)  # y-grid points

u = numpy.ones((ny, nx))  # Create a 2D array of ones for u
v = numpy.ones((ny, nx))  # Create a 2D array of ones for v
un = numpy.ones((ny, nx))  # Temporary array to hold values during updates for u
vn = numpy.ones((ny, nx))  # Temporary array to hold values during updates for v
comb = numpy.ones((ny, nx))  # Temporary array for combined values

### Assign initial conditions
# Set hat function initial condition: u(.5<=x<=1 && .5<=y<=1) is 2
u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
# Set hat function initial condition: v(.5<=x<=1 && .5<=y<=1) is 2
v[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2

### Plot Initial Conditions
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d'
X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting
ax.plot_surface(X, Y, u[:], cmap=cm.viridis, rstride=1, cstride=1)
ax.plot_surface(X, Y, v[:], cmap=cm.viridis, rstride=1, cstride=1)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

### Run through nt timesteps
for n in range(nt + 1):  # Loop across number of time steps
    un = u.copy()  # Copy the current state to temporary array for u
    vn = v.copy()  # Copy the current state to temporary array for v

    # Update u using finite difference method
    u[1:-1, 1:-1] = (un[1:-1, 1:-1] -
                     dt / dx * un[1:-1, 1:-1] * 
                     (un[1:-1, 1:-1] - un[1:-1, 0:-2]) - 
                     dt / dy * vn[1:-1, 1:-1] * 
                     (un[1:-1, 1:-1] - un[0:-2, 1:-1]) + 
                     nu * dt / dx**2 * 
                     (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, 0:-2]) + 
                     nu * dt / dy**2 * 
                     (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[0:-2, 1:-1]))

    # Update v using finite difference method
    v[1:-1, 1:-1] = (vn[1:-1, 1:-1] - 
                     dt / dx * un[1:-1, 1:-1] *
                     (vn[1:-1, 1:-1] - vn[1:-1, 0:-2]) -
                     dt / dy * vn[1:-1, 1:-1] * 
                     (vn[1:-1, 1:-1] - vn[0:-2, 1:-1]) + 
                     nu * dt / dx**2 * 
                     (vn[1:-1, 2:] - 2 * vn[1:-1, 1:-1] + vn[1:-1, 0:-2]) +
                     nu * dt / dy**2 *
                     (vn[2:, 1:-1] - 2 * vn[1:-1, 1:-1] + vn[0:-2, 1:-1]))
    
    # Apply boundary conditions
    u[0, :] = 1
    u[-1, :] = 1
    u[:, 0] = 1
    u[:, -1] = 1
    
    v[0, :] = 1
    v[-1, :] = 1
    v[:, 0] = 1
    v[:, -1] = 1

# Plot the result after nt timesteps
fig = pyplot.figure(figsize=(11, 7), dpi=100)
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot with projection='3d'
X, Y = numpy.meshgrid(x, y)
ax.plot_surface(X, Y, u, cmap=cm.viridis, rstride=1, cstride=1)
ax.plot_surface(X, Y, v, cmap=cm.viridis, rstride=1, cstride=1)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
pyplot.show()
```

Run Through nt Timesteps:
The main loop runs for nt time steps, updating u and v using finite differences to simulate diffusion and convection.
Boundary conditions are applied to ensure the edges of the grid remain fixed.

Final Plot:
The final state of u and v after nt time steps is plotted in 3D using ax.plot_surface.

# Step 9: 2D Laplace Equation

```python
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

def plot2D(x, y, p):
    """
    Function to plot a 3D surface plot of the 2D array p.

    Parameters:
    x (numpy.ndarray): 1D array of x coordinates.
    y (numpy.ndarray): 1D array of y coordinates.
    p (numpy.ndarray): 2D array of values to plot.

    """
    fig = pyplot.figure(figsize=(11, 7), dpi=100)  # Create a figure
    ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot
    X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting
    # Plot the surface
    surf = ax.plot_surface(X, Y, p[:], rstride=1, cstride=1, cmap=cm.viridis,
                           linewidth=0, antialiased=False)
    ax.set_xlim(0, 2)  # Set x-axis limits
    ax.set_ylim(0, 1)  # Set y-axis limits
    ax.view_init(30, 225)  # Set the view angle
    ax.set_xlabel('$x$')  # Set x-axis label
    ax.set_ylabel('$y$')  # Set y-axis label
    pyplot.show()  # Display the plot
```

```python
def laplace2d(p, y, dx, dy, l1norm_target):
    l1norm = 1
    pn = numpy.empty_like(p)

    while l1norm > l1norm_target:
        pn = p.copy()
        p[1:-1, 1:-1] = ((dy**2 * (pn[1:-1, 2:] + pn[1:-1, 0:-2]) +
                         dx**2 * (pn[2:, 1:-1] + pn[0:-2, 1:-1])) /
                        (2 * (dx**2 + dy**2)))
            
        p[:, 0] = 0  # p = 0 @ x = 0
        p[:, -1] = y  # p = y @ x = 2
        p[0, :] = p[1, :]  # dp/dy = 0 @ y = 0
        p[-1, :] = p[-2, :]  # dp/dy = 0 @ y = 1
        l1norm = (numpy.sum(numpy.abs(p[:]) - numpy.abs(pn[:])) /
                numpy.sum(numpy.abs(pn[:])))
     
    return p
```

```python
##variable declarations
nx = 31
ny = 31
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)


##initial conditions
p = numpy.zeros((ny, nx))  # create a XxY vector of 0's


##plotting aids
x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 1, ny)

##boundary conditions
p[:, 0] = 0  # p = 0 @ x = 0
p[:, -1] = y  # p = y @ x = 2
p[0, :] = p[1, :]  # dp/dy = 0 @ y = 0
p[-1, :] = p[-2, :]  # dp/dy = 0 @ y = 1


plot2D(x, y, p)
```

```python
p = laplace2d(p, y, dx, dy, 1e-4)
plot2D(x, y, p)
```

# Step 10: 2D Poisson Equation

```python
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

# Parameters
nx = 50  # Number of grid points in x direction
ny = 50  # Number of grid points in y direction
nt = 100  # Number of time steps
xmin, xmax = 0, 2  # Domain limits in x direction
ymin, ymax = 0, 1  # Domain limits in y direction

dx = (xmax - xmin) / (nx - 1)  # Grid spacing in x direction
dy = (ymax - ymin) / (ny - 1)  # Grid spacing in y direction

# Initialization
p = numpy.zeros((ny, nx))  # Pressure field
pd = numpy.zeros((ny, nx))  # Pressure field from previous time step
b = numpy.zeros((ny, nx))  # Source term
x = numpy.linspace(xmin, xmax, nx)  # x coordinates
y = numpy.linspace(ymin, ymax, ny)  # y coordinates

# Source term setup
b[int(ny / 4), int(nx / 4)] = 100  # Positive source at (ny/4, nx/4)
b[int(3 * ny / 4), int(3 * nx / 4)] = -100  # Negative source at (3*ny/4, 3*nx/4)

# Iterative solver
for it in range(nt):
    pd = p.copy()  # Copy the previous pressure field
    # Update pressure field using finite difference method
    p[1:-1, 1:-1] = (((pd[1:-1, 2:] + pd[1:-1, :-2]) * dy**2 +
                      (pd[2:, 1:-1] + pd[:-2, 1:-1]) * dx**2 -
                      b[1:-1, 1:-1] * dx**2 * dy**2) /
                     (2 * (dx**2 + dy**2)))

    # Boundary conditions
    p[0, :] = 0
    p[ny-1, :] = 0
    p[:, 0] = 0
    p[:, nx-1] = 0

# Function to plot the 2D pressure field
def plot2D(x, y, p):
    fig = pyplot.figure(figsize=(11, 7), dpi=100)  # Create a figure
    ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot
    X, Y = numpy.meshgrid(x, y)  # Create meshgrid for plotting
    # Plot the surface
    surf = ax.plot_surface(X, Y, p[:], rstride=1, cstride=1, cmap=cm.viridis,
                           linewidth=0, antialiased=False)
    ax.view_init(30, 225)  # Set the view angle
    ax.set_xlabel('$x$')  # Set x-axis label
    ax.set_ylabel('$y$')  # Set y-axis label
    pyplot.show()  # Display the plot

# Plot the pressure field
plot2D(x, y, p)
```

# Step 11 Solves the Navier-Stokes equation for 2D cavity flow

```python
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

# Grid parameters
nx = 41  # number of grid points in the x-direction
ny = 41  # number of grid points in the y-direction
nt = 500  # number of time steps
nit = 50  # number of iterations for the pressure Poisson equation
c = 1  # wave speed (not used here)
dx = 2 / (nx - 1)  # distance between grid points in the x-direction
dy = 2 / (ny - 1)  # distance between grid points in the y-direction
x = numpy.linspace(0, 2, nx)  # x grid points
y = numpy.linspace(0, 2, ny)  # y grid points
X, Y = numpy.meshgrid(x, y)  # create a meshgrid for plotting

# Physical parameters
rho = 1  # density
nu = 0.1  # kinematic viscosity
dt = 0.001  # time step

# Initial conditions
u = numpy.zeros((ny, nx))  # x-component of velocity
v = numpy.zeros((ny, nx))  # y-component of velocity
p = numpy.zeros((ny, nx))  # pressure field
b = numpy.zeros((ny, nx))  # source term

# Function to build up the source term b
def build_up_b(b, rho, dt, u, v, dx, dy):
    b[1:-1, 1:-1] = (rho * (1 / dt * 
                    ((u[1:-1, 2:] - u[1:-1, :-2]) / 
                     (2 * dx) + (v[2:, 1:-1] - v[:-2, 1:-1]) / (2 * dy)) -
                    ((u[1:-1, 2:] - u[1:-1, :-2]) / (2 * dx))**2 -
                      2 * ((u[2:, 1:-1] - u[:-2, 1:-1]) / (2 * dy) *
                           (v[1:-1, 2:] - v[1:-1, :-2]) / (2 * dx))-
                          ((v[2:, 1:-1] - v[:-2, 1:-1]) / (2 * dy))**2))
    return b

# Function to solve the pressure Poisson equation
def pressure_poisson(p, dx, dy, b):
    pn = numpy.empty_like(p)
    for q in range(nit):
        pn = p.copy()
        p[1:-1, 1:-1] = (((pn[1:-1, 2:] + pn[1:-1, :-2]) * dy**2 + 
                          (pn[2:, 1:-1] + pn[:-2, 1:-1]) * dx**2) /
                          (2 * (dx**2 + dy**2)) -
                          dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * 
                          b[1:-1,1:-1])

        # Boundary conditions
        p[:, -1] = p[:, -2]  # dp/dy = 0 at x = 2
        p[0, :] = p[1, :]  # dp/dy = 0 at y = 0
        p[:, 0] = p[:, 1]  # dp/dx = 0 at x = 0
        p[-1, :] = 0  # p = 0 at y = 2
        
    return p

# Function to solve the Navier-Stokes equations
def cavity_flow(nt, u, v, dt, dx, dy, p, rho, nu):
    un = numpy.empty_like(u)
    vn = numpy.empty_like(v)
    b = numpy.zeros((ny, nx))
    
    for n in range(nt):
        un = u.copy()
        vn = v.copy()
        
        b = build_up_b(b, rho, dt, u, v, dx, dy)
        p = pressure_poisson(p, dx, dy, b)
        
        u[1:-1, 1:-1] = (un[1:-1, 1:-1] -
                         un[1:-1, 1:-1] * dt / dx *
                        (un[1:-1, 1:-1] - un[1:-1, :-2]) -
                         vn[1:-1, 1:-1] * dt / dy *
                        (un[1:-1, 1:-1] - un[:-2, 1:-1]) -
                         dt / (2 * rho * dx) * (p[1:-1, 2:] - p[1:-1, :-2]) +
                         nu * (dt / dx**2 *
                        (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, :-2]) +
                         dt / dy**2 *
                        (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[:-2, 1:-1])))

        v[1:-1, 1:-1] = (vn[1:-1, 1:-1] -
                         un[1:-1, 1:-1] * dt / dx *
                        (vn[1:-1, 1:-1] - vn[1:-1, :-2]) -
                         vn[1:-1, 1:-1] * dt / dy *
                        (vn[1:-1, 1:-1] - vn[:-2, 1:-1]) -
                         dt / (2 * rho * dy) * (p[2:, 1:-1] - p[:-2, 1:-1]) +
                         nu * (dt / dx**2 *
                        (vn[1:-1, 2:] - 2 * vn[1:-1, 1:-1] + vn[1:-1, :-2]) +
                         dt / dy**2 *
                        (vn[2:, 1:-1] - 2 * vn[1:-1, 1:-1] + vn[:-2, 1:-1])))

        # Boundary conditions for velocity
        u[0, :] = 0
        u[:, 0] = 0
        u[:, -1] = 0
        u[-1, :] = 1  # set velocity on cavity lid equal to 1
        v[0, :] = 0
        v[-1, :] = 0
        v[:, 0] = 0
        v[:, -1] = 0
        
    return u, v, p

# Initial conditions
u = numpy.zeros((ny, nx))
v = numpy.zeros((ny, nx))
p = numpy.zeros((ny, nx))
b = numpy.zeros((ny, nx))

# Solve for the velocity and pressure fields
nt = 100
u, v, p = cavity_flow(nt, u, v, dt, dx, dy, p, rho, nu)

# Plotting the results
fig = pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)  # pressure field as contour
pyplot.colorbar()
pyplot.contour(X, Y, p, cmap=cm.viridis)  # pressure field outlines
pyplot.quiver(X[::2, ::2], Y[::2, ::2], u[::2, ::2], v[::2, ::2])  # velocity field
pyplot.xlabel('X')
pyplot.ylabel('Y')
pyplot.show()
```

```python
# Re-initialize and run for a longer time
u = numpy.zeros((ny, nx))
v = numpy.zeros((ny, nx))
p = numpy.zeros((ny, nx))
b = numpy.zeros((ny, nx))

nt = 700
u, v, p = cavity_flow(nt, u, v, dt, dx, dy, p, rho, nu)

# Plotting the results after longer time
fig = pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
pyplot.colorbar()
pyplot.contour(X, Y, p, cmap=cm.viridis)
pyplot.quiver(X[::2, ::2], Y[::2, ::2], u[::2, ::2], v[::2, ::2])
pyplot.xlabel('X')
pyplot.ylabel('Y')
pyplot.show()
```

# Step 12 Solves the Navier-Stokes equation for 2D channel flow

```python
import numpy
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

def build_up_b(rho, dt, dx, dy, u, v):
    b = numpy.zeros_like(u)
    b[1:-1, 1:-1] = (rho * (1 / dt * ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx) +
                                      (v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy)) -
                            ((u[1:-1, 2:] - u[1:-1, 0:-2]) / (2 * dx))**2 -
                            2 * ((u[2:, 1:-1] - u[0:-2, 1:-1]) / (2 * dy) *
                                 (v[1:-1, 2:] - v[1:-1, 0:-2]) / (2 * dx))-
                            ((v[2:, 1:-1] - v[0:-2, 1:-1]) / (2 * dy))**2))
    
    # Periodic BC Pressure @ x = 2
    b[1:-1, -1] = (rho * (1 / dt * ((u[1:-1, 0] - u[1:-1,-2]) / (2 * dx) +
                                    (v[2:, -1] - v[0:-2, -1]) / (2 * dy)) -
                          ((u[1:-1, 0] - u[1:-1, -2]) / (2 * dx))**2 -
                          2 * ((u[2:, -1] - u[0:-2, -1]) / (2 * dy) *
                               (v[1:-1, 0] - v[1:-1, -2]) / (2 * dx)) -
                          ((v[2:, -1] - v[0:-2, -1]) / (2 * dy))**2))

    # Periodic BC Pressure @ x = 0
    b[1:-1, 0] = (rho * (1 / dt * ((u[1:-1, 1] - u[1:-1, -1]) / (2 * dx) +
                                   (v[2:, 0] - v[0:-2, 0]) / (2 * dy)) -
                         ((u[1:-1, 1] - u[1:-1, -1]) / (2 * dx))**2 -
                         2 * ((u[2:, 0] - u[0:-2, 0]) / (2 * dy) *
                              (v[1:-1, 1] - v[1:-1, -1]) / (2 * dx))-
                         ((v[2:, 0] - v[0:-2, 0]) / (2 * dy))**2))
    
    return b
```

```python
def pressure_poisson_periodic(p, dx, dy):
    pn = numpy.empty_like(p)

    for q in range(nit):
        pn = p.copy()
        p[1:-1, 1:-1] = (((pn[1:-1, 2:] + pn[1:-1, 0:-2]) * dy**2 +
                          (pn[2:, 1:-1] + pn[0:-2, 1:-1]) * dx**2) /
                         (2 * (dx**2 + dy**2)) -
                         dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, 1:-1])

        # Periodic BC Pressure @ x = 2
        p[1:-1, -1] = (((pn[1:-1, 0] + pn[1:-1, -2])* dy**2 +
                        (pn[2:, -1] + pn[0:-2, -1]) * dx**2) /
                       (2 * (dx**2 + dy**2)) -
                       dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, -1])

        # Periodic BC Pressure @ x = 0
        p[1:-1, 0] = (((pn[1:-1, 1] + pn[1:-1, -1])* dy**2 +
                       (pn[2:, 0] + pn[0:-2, 0]) * dx**2) /
                      (2 * (dx**2 + dy**2)) -
                      dx**2 * dy**2 / (2 * (dx**2 + dy**2)) * b[1:-1, 0])
        
        # Wall boundary conditions, pressure
        p[-1, :] =p[-2, :]  # dp/dy = 0 at y = 2
        p[0, :] = p[1, :]  # dp/dy = 0 at y = 0
    
    return p

```

```python
##variable declarations
nx = 41
ny = 41
nt = 10
nit = 50 
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
x = numpy.linspace(0, 2, nx)
y = numpy.linspace(0, 2, ny)
X, Y = numpy.meshgrid(x, y)

##physical variables
rho = 1
nu = .1
F = 1
dt = .01

#initial conditions
u = numpy.zeros((ny, nx))
un = numpy.zeros((ny, nx))

v = numpy.zeros((ny, nx))
vn = numpy.zeros((ny, nx))

p = numpy.ones((ny, nx))
pn = numpy.ones((ny, nx))

b = numpy.zeros((ny, nx))
```

```python
udiff = 1
stepcount = 0

while udiff > .001:
    un = u.copy()
    vn = v.copy()

    b = build_up_b(rho, dt, dx, dy, u, v)
    p = pressure_poisson_periodic(p, dx, dy)

    u[1:-1, 1:-1] = (un[1:-1, 1:-1] -
                     un[1:-1, 1:-1] * dt / dx * 
                    (un[1:-1, 1:-1] - un[1:-1, 0:-2]) -
                     vn[1:-1, 1:-1] * dt / dy * 
                    (un[1:-1, 1:-1] - un[0:-2, 1:-1]) -
                     dt / (2 * rho * dx) * 
                    (p[1:-1, 2:] - p[1:-1, 0:-2]) +
                     nu * (dt / dx**2 * 
                    (un[1:-1, 2:] - 2 * un[1:-1, 1:-1] + un[1:-1, 0:-2]) +
                     dt / dy**2 * 
                    (un[2:, 1:-1] - 2 * un[1:-1, 1:-1] + un[0:-2, 1:-1])) + 
                     F * dt)

    v[1:-1, 1:-1] = (vn[1:-1, 1:-1] -
                     un[1:-1, 1:-1] * dt / dx * 
                    (vn[1:-1, 1:-1] - vn[1:-1, 0:-2]) -
                     vn[1:-1, 1:-1] * dt / dy * 
                    (vn[1:-1, 1:-1] - vn[0:-2, 1:-1]) -
                     dt / (2 * rho * dy) * 
                    (p[2:, 1:-1] - p[0:-2, 1:-1]) +
                     nu * (dt / dx**2 *
                    (vn[1:-1, 2:] - 2 * vn[1:-1, 1:-1] + vn[1:-1, 0:-2]) +
                     dt / dy**2 * 
                    (vn[2:, 1:-1] - 2 * vn[1:-1, 1:-1] + vn[0:-2, 1:-1])))

   # Periodic BC u @ x = 2     
    u[1:-1, -1] = (un[1:-1, -1] - un[1:-1, -1] * dt / dx * 
                  (un[1:-1, -1] - un[1:-1, -2]) -
                   vn[1:-1, -1] * dt / dy * 
                  (un[1:-1, -1] - un[0:-2, -1]) -
                   dt / (2 * rho * dx) *
                  (p[1:-1, 0] - p[1:-1, -2]) + 
                   nu * (dt / dx**2 * 
                  (un[1:-1, 0] - 2 * un[1:-1,-1] + un[1:-1, -2]) +
                   dt / dy**2 * 
                  (un[2:, -1] - 2 * un[1:-1, -1] + un[0:-2, -1])) + F * dt)

    # Periodic BC u @ x = 0
    u[1:-1, 0] = (un[1:-1, 0] - un[1:-1, 0] * dt / dx *
                 (un[1:-1, 0] - un[1:-1, -1]) -
                  vn[1:-1, 0] * dt / dy * 
                 (un[1:-1, 0] - un[0:-2, 0]) - 
                  dt / (2 * rho * dx) * 
                 (p[1:-1, 1] - p[1:-1, -1]) + 
                  nu * (dt / dx**2 * 
                 (un[1:-1, 1] - 2 * un[1:-1, 0] + un[1:-1, -1]) +
                  dt / dy**2 *
                 (un[2:, 0] - 2 * un[1:-1, 0] + un[0:-2, 0])) + F * dt)

    # Periodic BC v @ x = 2
    v[1:-1, -1] = (vn[1:-1, -1] - un[1:-1, -1] * dt / dx *
                  (vn[1:-1, -1] - vn[1:-1, -2]) - 
                   vn[1:-1, -1] * dt / dy *
                  (vn[1:-1, -1] - vn[0:-2, -1]) -
                   dt / (2 * rho * dy) * 
                  (p[2:, -1] - p[0:-2, -1]) +
                   nu * (dt / dx**2 *
                  (vn[1:-1, 0] - 2 * vn[1:-1, -1] + vn[1:-1, -2]) +
                   dt / dy**2 *
                  (vn[2:, -1] - 2 * vn[1:-1, -1] + vn[0:-2, -1])))

    # Periodic BC v @ x = 0
    v[1:-1, 0] = (vn[1:-1, 0] - un[1:-1, 0] * dt / dx *
                 (vn[1:-1, 0] - vn[1:-1, -1]) -
                  vn[1:-1, 0] * dt / dy *
                 (vn[1:-1, 0] - vn[0:-2, 0]) -
                  dt / (2 * rho * dy) * 
                 (p[2:, 0] - p[0:-2, 0]) +
                  nu * (dt / dx**2 * 
                 (vn[1:-1, 1] - 2 * vn[1:-1, 0] + vn[1:-1, -1]) +
                  dt / dy**2 * 
                 (vn[2:, 0] - 2 * vn[1:-1, 0] + vn[0:-2, 0])))


    # Wall BC: u,v = 0 @ y = 0,2
    u[0, :] = 0
    u[-1, :] = 0
    v[0, :] = 0
    v[-1, :]=0
    
    udiff = (numpy.sum(u) - numpy.sum(un)) / numpy.sum(u)
    stepcount += 1
#You can see that we've also included a variable stepcount to see how many iterations our loop went through before our stop condition was met.

fig = pyplot.figure(figsize = (11,7), dpi=100)
pyplot.quiver(X[::3, ::3], Y[::3, ::3], u[::3, ::3], v[::3, ::3]);
pyplot.show();
```
